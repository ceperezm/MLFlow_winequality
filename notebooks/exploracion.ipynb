```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import arff

# Cargar datos
data, meta = arff.loadarff('../data/winequality-red.csv.arff')

# Convertir a DataFrame
df = pd.DataFrame(data)


```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df['quality'].value_counts())
```

```python

# Contar valores no nulos por columna
conteo = df.count().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=conteo.index, y=conteo.values, palette='Set2')

plt.title('Cantidad de valores por columna')
plt.xlabel('Columnas')
plt.ylabel('Cantidad de valores (no nulos)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlacin entre variables')
plt.show()
```

```python
correlation = df.corr()['quality'].sort_values(ascending=False)
print(correlation)
```
